```python
#Exploring the Relationship Between Temperature and Sales of Ice Creams
```

```python
import pandas as pd # data processing, CSV file I/O
import matplotlib.pyplot as plt # plots, export to many file formats
```

```python
df = pd.read_csv('IceCreamData.csv')
df
```

```python
result = df.sort_values('Temperature')
result
```

```python
df.columns
```

```python
df.rename(columns={'Temperature':'Temperature (Celsius)'},inplace=True)
```

```python
df.head()
```

```python
plt.plot(result['Temperature'], result['Revenue'], 'r.-', label= 'US$')
plt.title('Ice Cream Revenues')
plt.ylabel('Sales (US$)')
plt.xlabel('Temperature (Celsius)')
plt.legend()
plt.rcParams['figure.figsize']=(20,15)

plt.savefig('IceCream_revenues.png')
plt.show()
```
